```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/crypto_data_updated_13_november.csv", sep=',')
drop= df.drop(['Date', 'Volume (BTC)','Volume (ETH)','Volume (USDT)','Volume (BNB)'], axis=1)
sns.set_style('darkgrid')
sns.set(font_scale=1.3)
sns.pairplot(drop, aspect=.85)
```

```python
df = pd.read_csv("/content/crypto_data_updated_13_november.csv", sep=',')
drop= df.drop(['Date', 'Volume (BTC)','Volume (ETH)','Volume (USDT)','Volume (BNB)'], axis=1)
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(drop.corr(), xticklabels=drop.corr().columns,
 yticklabels=drop.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
